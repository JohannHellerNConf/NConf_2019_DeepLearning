# **Download Dataset**

Das Hunde und Katzen Datenset ist ein Beispiel für eine Computer-Vision Challengen welche sich besonders gut mit Deep Learning lösen lässt. Für genauere Informationen: https://www.kaggle.com/c/dogs-vs-cats

Dieses Script downloaded das Katzen-und-Hunde Datenset aus einem AWS S3 Bucket und entpackt es anschließend.

```
!wget -c https://s3.amazonaws.com/content.udacity-data.com/nd089/Cat_Dog_data.zip

!unzip -qq Cat_Dog_data.zip
```

Wir können nach der Ausführung die heruntergeladenen Dateien einsehen:

```
!ls -l

!ls -l Cat_Dog_data/
```
